```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
confirmed = pd.read_csv(r"D:\Projects\CoronaVirusAnalysis\FormattedData\confirmed.csv",index_col=0)
deaths = pd.read_csv(r"D:\Projects\CoronaVirusAnalysis\FormattedData\deaths.csv",index_col=0)
recovered = pd.read_csv(r"D:\Projects\CoronaVirusAnalysis\FormattedData\recovered.csv",index_col=0)
```

```python
new_cases = confirmed.copy()

for day in range(1,len(confirmed)):
    new_cases.iloc[day] = confirmed.iloc[day] - confirmed.iloc[day-1]
```

```python
growth_rate = confirmed.copy()

for day in range(1,len(confirmed)):
    growth_rate.iloc[day] = (new_cases.iloc[day] / confirmed.iloc[day-1] * 100)
```
