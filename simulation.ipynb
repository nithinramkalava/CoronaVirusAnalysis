```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
simulation_growth_rate = 0.1

confirmed = pd.read_csv("FormattedData\confirmed.csv",index_col=0)

dates = pd.date_range(start='2023-01-20', periods=40, freq='D')
dates = pd.Series(dates)
dates = dates.dt.strftime('%d/%m/%Y')

simulated = confirmed.copy()
simulated = simulated.append(pd.DataFrame(index=dates))
```

```python
plt.figure(figsize=(15,6),dpi=200)

ax = simulated['India'][10:].plot(label="Italy")
ax.set_title('COVID-19 India')
```
