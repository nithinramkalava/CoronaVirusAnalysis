```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
confirmed = pd.read_csv("FormattedData\confirmed.csv",index_col=0)
```

```python
countries = "India,US,China,Spain,France,Sweden,Bangladesh"

plt.figure(figsize=(15,6),dpi=200)

for country in countries.split(","):
    confirmed[country].plot(label=country)

plt.legend(loc="upper left")
```

```python
growth_rate = pd.read_csv("Rates/growth_rate.csv",index_col=0)
```

```python
plt.figure(figsize=(15,6),dpi=200)
plt.xticks(rotation=90,size=2)
plt.xlabel("Timeline --->")
plt.title("Growth Rate of Covid-19 since february 2022 - India")

sns.barplot(x = growth_rate[750:].index, y = growth_rate.India[750:])
plt.ylabel("India Covid-19 Growth Rate")
```

```python
death_rate = pd.read_csv("Rates/death_rate.csv",index_col=0)
```

```python
deaths = pd.read_csv("FormattedData/deaths.csv",index_col=0)
plt.figure(figsize=(15,6),dpi=200)
plt.xticks(rotation=90,size=1)
plt.title("Deaths due to Covid-19 since January 2020 - India")

sns.barplot(x=deaths.index, y=deaths.India)
```

```python
plt.figure(figsize=(15,6),dpi=200)
plt.xticks(rotation=90,size=2)
plt.xlabel("Timeline --->")
plt.title("Death Rate of Covid-19 since January 2020 - India")

sns.barplot(x = death_rate.index, y = death_rate.India)
plt.ylabel("India Covid-19 Death Rate")
```
